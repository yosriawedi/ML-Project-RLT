# **Business Understading**

**Problem Statement**
many real-world applications, datasets contain a very large number of variables (p can reach hundreds or even thousands), while the number of available observations remains relatively small. In such situations, only a small subset of variables actually carries meaningful information for predicting the target, whereas the remaining variables act merely as noise.

This high-dimensional setting creates several challenges for traditional machine learning models. Methods such as Random Forests, CART, and other standard tree-based algorithms often fail to correctly identify the truly informative variables, avoid selecting irrelevant features during the splitting process, and capture complex interactions between variables when these interactions do not manifest through marginal effects.

Therefore, the central question is:

How can we build a predictive model that remains accurate, stable, and reliable in high-dimensional environments (“large p, small n”), where only a small portion of the variables contains real signal, classical models struggle to detect important features, and crucial interactions may exist without any marginal effect?


**BO1 — Identify what truly matters within the available information**

Many pieces of information are collected, but only a small part is genuinely useful for understanding or predicting the outcome. The goal is to determine which elements are truly important and which ones do not contribute meaningfully

DSO1
Build a model capable of reliably estimating variable importance in order to distinguish strong variables from noise variables, even when p ≫ n.

**BO2 Improve the quality of the system’s decision-making process **

Ensure that the decision-support tool relies mainly on meaningful and relevant information, instead of being influenced by random or misleading elements, so that the results become more consistent and trustworthy..

DSO2
Use optimization mechanisms such as reinforcement learning and variable muting to guide the model’s splits toward the variables that are truly relevant.

 **BO3
Ensure that the system remains reliable even with a lot of information but few case”**
The organization needs a predictive tool that stays stable and dependable, even when there are many different pieces of information per case but only a limited number of cases available.

DSO3
Develop a predictive model that remains statistically reliable even when the number of variables is significantly higher than the number of observations.

```python
# ============================================================
# DATA PREPARATION (CRISP-DM) — Auto MPG + Ozone
# 1 seul script Colab pour les deux datasets
# ============================================================

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# -------------------------------
# 0) Chemins (modifie si besoin)
# -------------------------------
AUTO_PATH  = "/content/auto-mpg.data"
OZONE_PATH = "/content/onehr.data"   # ou "/content/eighthr.data"
OZONE_VARIANT = "onehr"             # "onehr" ou "eighthr"

TEST_SIZE = 0.30
SEED = 42

# ============================================================
# 1) AUTO MPG — Chargement + Préparation
# ============================================================

AUTO_COLS = [
    "mpg","cylinders","displacement","horsepower","weight",
    "acceleration","model_year","origin","car_name"
]

def load_auto_mpg(path):
    df = pd.read_csv(
        path,
        delim_whitespace=True,
        header=None,
        names=AUTO_COLS,
        na_values=["?"]
    )
    return df

def prepare_auto_mpg(df, drop_car_name=True):
    # Target
    y = df["mpg"].astype(float)
    X = df.drop(columns=["mpg"])

    # supprimer texte inutile
    if drop_car_name and "car_name" in X.columns:
        X = X.drop(columns=["car_name"])

    # Colonnes catégorielles / numériques
    cat_cols = ["origin"] if "origin" in X.columns else []
    num_cols = [c for c in X.columns if c not in cat_cols]

    # Pipelines CRISP-DM
    numeric_pipe = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ])

    categorical_pipe = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore"))
    ])

    preprocessor = ColumnTransformer(
        transformers=[
            ("num", numeric_pipe, num_cols),
            ("cat", categorical_pipe, cat_cols)
        ],
        remainder="drop"
    )

    # Split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=TEST_SIZE, random_state=SEED
    )

    # Fit/transform train ; transform test
    X_train_p = preprocessor.fit_transform(X_train)
    X_test_p  = preprocessor.transform(X_test)

    return X_train_p, X_test_p, y_train.to_numpy(), y_test.to_numpy()


# ============================================================
# 2) OZONE — Chargement + Préparation
# ============================================================

def load_ozone(path):
    df = pd.read_csv(path, header=None)
    return df

def prepare_ozone(df):
    # target = dernière colonne
    y = pd.to_numeric(df.iloc[:, -1], errors="coerce")

    # features = colonnes 1..-2 (col0=date retirée)
    X = df.iloc[:, 1:-1].apply(pd.to_numeric, errors="coerce")

    preprocessor = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ])

    # Split stratifié (classe rare)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=TEST_SIZE, random_state=SEED, stratify=y
    )

    X_train_p = preprocessor.fit_transform(X_train)
    X_test_p  = preprocessor.transform(X_test)

    return X_train_p, X_test_p, y_train.to_numpy(), y_test.to_numpy()


# ============================================================
# 3) OPTION (DSO) — Ajouter du bruit pour simuler high-dim
# ============================================================

def add_noise_features(X_train, X_test, p_total=500, seed=42):
    """
    Ajoute des variables bruitées N(0,1) pour atteindre p_total variables.
    Utile pour tester RLT dans un cadre sparse high-dim.
    """
    rng = np.random.default_rng(seed)
    n_train, p_curr = X_train.shape
    n_test  = X_test.shape[0]

    if p_total <= p_curr:
        return X_train, X_test  # rien à ajouter

    p_noise = p_total - p_curr
    noise_train = rng.normal(0, 1, size=(n_train, p_noise))
    noise_test  = rng.normal(0, 1, size=(n_test,  p_noise))

    X_train_hd = np.hstack([X_train, noise_train])
    X_test_hd  = np.hstack([X_test,  noise_test ])

    return X_train_hd, X_test_hd


# ============================================================
# 4) EXÉCUTION
# ============================================================

# ---- Auto MPG ----
auto_df = load_auto_mpg(AUTO_PATH)
X_train_auto, X_test_auto, y_train_auto, y_test_auto = prepare_auto_mpg(auto_df)

print("Auto MPG prepared:")
print("  X_train:", X_train_auto.shape, "X_test:", X_test_auto.shape)

# ---- Ozone ----
ozone_df = load_ozone(OZONE_PATH)
X_train_ozone, X_test_ozone, y_train_ozone, y_test_ozone = prepare_ozone(ozone_df)

print("\nOzone prepared (" + OZONE_VARIANT + "):")
print("  X_train:", X_train_ozone.shape, "X_test:", X_test_ozone.shape)
print("  Class train distribution:", np.bincount(y_train_ozone.astype(int)))

# (Option) activer pour simuler un vrai large-p
# X_train_auto, X_test_auto = add_noise_features(X_train_auto, X_test_auto, p_total=500, seed=SEED)
# X_train_ozone, X_test_ozone = add_noise_features(X_train_ozone, X_test_ozone, p_total=500, seed=SEED)

# Maintenant tu peux passer au Modeling (RF/ExtraTrees/RLT)
```

```python
# ============================================================
# Étape B — Reinforcement Learning Trees (RLT) en Python
# Reproduction pédagogique basée sur Zhu, Zeng & Kosorok (2015)
# ============================================================

import numpy as np
from dataclasses import dataclass
from sklearn.ensemble import ExtraTreesRegressor, ExtraTreesClassifier

# ------------------------------
# Utils pertes
# ------------------------------
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def logloss(y_true, proba, eps=1e-12):
    proba = np.clip(proba, eps, 1 - eps)
    # y_true in {0,1}
    return -np.mean(y_true * np.log(proba) + (1 - y_true) * np.log(1 - proba))

# ------------------------------
# Noeud d'arbre RLT
# ------------------------------
@dataclass
class RLTNode:
    is_leaf: bool
    value: any = None                 # mean (regression) or class prob (classification)
    feature: int = None               # index feature for RLT1
    threshold: float = None
    beta: np.ndarray = None           # linear-combination weights for RLT2+
    left: "RLTNode" = None
    right: "RLTNode" = None

class RLTTree:
    """
    Un arbre RLT.
    - embedded ExtraTrees pour estimer VI locale au noeud
    - muting / protection
    - split RLT1 (variable max VI)
    - split RLT2+ (combinaison lineaire top-k)
    """
    def __init__(
        self,
        task="regression",             # "regression" or "classification"
        min_samples_leaf=5,            # nmin (taille minimale feuille)
        pd_frac=0.5,                   # fraction de variables mutées par niveau
        p0_min_active=0,               # p0: nb min de variables actives
        k_lincomb=1,                   # k=1 => RLT1 ; k>1 => RLT2+
        alpha_vi=0.25,                 # seuil alpha (variables assez importantes)
        embedded_estimators=50,        # nb d'arbres ExtraTrees embarqués
        embedded_max_features="sqrt",  # comme ExtraTrees
        n_thresholds=10,               # nb seuils random testés (style ExtraTrees)
        random_state=None
    ):
        self.task = task
        self.min_samples_leaf = min_samples_leaf
        self.pd_frac = pd_frac
        self.p0_min_active = p0_min_active
        self.k_lincomb = k_lincomb
        self.alpha_vi = alpha_vi
        self.embedded_estimators = embedded_estimators
        self.embedded_max_features = embedded_max_features
        self.n_thresholds = n_thresholds
        self.rng = np.random.default_rng(random_state)

        self.root = None
        self.n_features_ = None

    # --------------------------
    # Embedded model au noeud
    # --------------------------
    def _fit_embedded(self, X, y, active_features):
        Xa = X[:, active_features]
        if self.task == "regression":
            model = ExtraTreesRegressor(
                n_estimators=self.embedded_estimators,
                max_features=self.embedded_max_features,
                random_state=self.rng.integers(0, 10**9),
                n_jobs=-1
            )
            model.fit(Xa, y)
            pred = model.predict(Xa)
            base_loss = mse(y, pred)
        else:
            model = ExtraTreesClassifier(
                n_estimators=self.embedded_estimators,
                max_features=self.embedded_max_features,
                random_state=self.rng.integers(0, 10**9),
                n_jobs=-1
            )
            model.fit(Xa, y)
            proba = model.predict_proba(Xa)[:, 1]
            base_loss = logloss(y, proba)

        return model, base_loss

    # --------------------------
    # VI par permutation (locale)
    # --------------------------
    def _permutation_vi(self, model, X, y, active_features, base_loss):
        Xa = X[:, active_features].copy()
        vi = np.zeros(len(active_features))

        for idx, f in enumerate(active_features):
            Xa_perm = Xa.copy()
            col = idx  # colonne dans Xa
            self.rng.shuffle(Xa_perm[:, col])

            if self.task == "regression":
                pred_p = model.predict(Xa_perm)
                loss_p = mse(y, pred_p)
            else:
                proba_p = model.predict_proba(Xa_perm)[:, 1]
                loss_p = logloss(y, proba_p)

            vi[idx] = max(0.0, loss_p - base_loss)  # gain de perte

        return vi

    # --------------------------
    # Choix split RLT1 (1 variable)
    # --------------------------
    def _best_split_single(self, X, y, active_features, vi):
        # variable avec VI max
        best_local_idx = np.argmax(vi)
        best_feat = active_features[best_local_idx]

        xcol = X[:, best_feat]
        xmin, xmax = np.min(xcol), np.max(xcol)
        if xmin == xmax:
            return None  # pas de split possible

        # tester plusieurs seuils aléatoires (ExtraTrees style)
        best_thr, best_score = None, np.inf
        for _ in range(self.n_thresholds):
            thr = self.rng.uniform(xmin, xmax)
            left_mask = xcol <= thr
            right_mask = ~left_mask

            if left_mask.sum() < self.min_samples_leaf or right_mask.sum() < self.min_samples_leaf:
                continue

            if self.task == "regression":
                score = (left_mask.sum() * np.var(y[left_mask]) +
                         right_mask.sum() * np.var(y[right_mask]))
            else:
                # Gini impurity pondérée
                def gini(yy):
                    p1 = np.mean(yy) if len(yy) else 0
                    return 2*p1*(1-p1)
                score = (left_mask.sum() * gini(y[left_mask]) +
                         right_mask.sum() * gini(y[right_mask]))

            if score < best_score:
                best_score, best_thr = score, thr

        if best_thr is None:
            return None
        return best_feat, best_thr

    # --------------------------
    # Choix split RLT2+ (combinaison lineaire top-k)
    # --------------------------
    def _best_split_linear(self, X, y, active_features, vi):
        # sélection des top-k avec VI assez grande (>= alpha * maxVI)
        max_vi = vi.max() if vi.size else 0.0
        if max_vi == 0.0:
            return None

        eligible = np.where(vi >= self.alpha_vi * max_vi)[0]
        if eligible.size == 0:
            return None

        # prendre top-k parmi éligibles
        eligible_sorted = eligible[np.argsort(vi[eligible])[::-1]]
        topk_local = eligible_sorted[: self.k_lincomb]
        topk_feats = [active_features[i] for i in topk_local]

        # beta_j = sign(corr(Xj, y)) * sqrt(VI_j)
        beta = np.zeros(self.n_features_)
        for local_i, f in zip(topk_local, topk_feats):
            xj = X[:, f]
            corr = np.corrcoef(xj, y)[0, 1]
            sgn = 1.0 if np.isnan(corr) or corr >= 0 else -1.0
            beta[f] = sgn * np.sqrt(vi[local_i] + 1e-12)

        # projection
        z = X @ beta
        zmin, zmax = np.min(z), np.max(z)
        if zmin == zmax:
            return None

        best_thr, best_score = None, np.inf
        for _ in range(self.n_thresholds):
            thr = self.rng.uniform(zmin, zmax)
            left_mask = z <= thr
            right_mask = ~left_mask

            if left_mask.sum() < self.min_samples_leaf or right_mask.sum() < self.min_samples_leaf:
                continue

            if self.task == "regression":
                score = (left_mask.sum() * np.var(y[left_mask]) +
                         right_mask.sum() * np.var(y[right_mask]))
            else:
                def gini(yy):
                    p1 = np.mean(yy) if len(yy) else 0
                    return 2*p1*(1-p1)
                score = (left_mask.sum() * gini(y[left_mask]) +
                         right_mask.sum() * gini(y[right_mask]))

            if score < best_score:
                best_score, best_thr = score, thr

        if best_thr is None:
            return None
        return beta, best_thr

    # --------------------------
    # Construction récursive
    # --------------------------
    def _build(self, X, y, active_features, protected_features):
        n = X.shape[0]

        # arrêt
        if n < 2*self.min_samples_leaf or len(active_features) == 0:
            return self._make_leaf(y)

        if self.task == "regression" and np.all(y == y[0]):
            return self._make_leaf(y)

        if self.task == "classification" and (y.mean() in [0.0, 1.0]):
            return self._make_leaf(y)

        # embedded model + VI
        emb_model, base_loss = self._fit_embedded(X, y, active_features)
        vi_local = self._permutation_vi(emb_model, X, y, active_features, base_loss)

        # muting: mute pd_frac plus faibles VI parmi NON protégées
        unprotected_idx = [i for i,f in enumerate(active_features) if f not in protected_features]
        if len(unprotected_idx) > 0:
            # nombre à muter
            pd_count = int(np.floor(self.pd_frac * len(unprotected_idx)))
            # respecter p0_min_active
            max_mute_allowed = max(0, len(active_features) - self.p0_min_active)
            pd_count = min(pd_count, max_mute_allowed)

            if pd_count > 0:
                unprot_vi = [(i, vi_local[i]) for i in unprotected_idx]
                unprot_vi.sort(key=lambda t: t[1])  # asc
                to_mute_local = [i for i,_ in unprot_vi[:pd_count]]
                to_mute_feats = {active_features[i] for i in to_mute_local}
                active_features = [f for f in active_features if f not in to_mute_feats]
                # recalculer vi_local sur nouveau active_features
                keep_mask = [i for i,f in enumerate(active_features)]
                # (simple: refit embedded après muting)
                emb_model, base_loss = self._fit_embedded(X, y, active_features)
                vi_local = self._permutation_vi(emb_model, X, y, active_features, base_loss)

        # choisir split
        if self.k_lincomb <= 1:
            split = self._best_split_single(X, y, active_features, vi_local)
            if split is None:
                return self._make_leaf(y)
            best_feat, best_thr = split

            # protection update
            new_protected = set(protected_features)
            new_protected.add(best_feat)

            xcol = X[:, best_feat]
            left_mask = xcol <= best_thr
            right_mask = ~left_mask

            left_node = self._build(X[left_mask], y[left_mask], active_features, new_protected)
            right_node = self._build(X[right_mask], y[right_mask], active_features, new_protected)

            return RLTNode(
                is_leaf=False, feature=best_feat, threshold=best_thr,
                left=left_node, right=right_node
            )

        else:
            split = self._best_split_linear(X, y, active_features, vi_local)
            if split is None:
                # fallback simple
                split = self._best_split_single(X, y, active_features, vi_local)
                if split is None:
                    return self._make_leaf(y)
                best_feat, best_thr = split
                new_protected = set(protected_features); new_protected.add(best_feat)
                xcol = X[:, best_feat]
                left_mask = xcol <= best_thr
                right_mask = ~left_mask
                left_node = self._build(X[left_mask], y[left_mask], active_features, new_protected)
                right_node = self._build(X[right_mask], y[right_mask], active_features, new_protected)
                return RLTNode(False, feature=best_feat, threshold=best_thr, left=left_node, right=right_node)

            beta, best_thr = split
            z = X @ beta
            left_mask = z <= best_thr
            right_mask = ~left_mask
            if left_mask.sum() < self.min_samples_leaf or right_mask.sum() < self.min_samples_leaf:
                return self._make_leaf(y)

            # variables utilisées dans beta deviennent protégées
            used_feats = set(np.where(beta != 0)[0])
            new_protected = set(protected_features).union(used_feats)

            left_node = self._build(X[left_mask], y[left_mask], active_features, new_protected)
            right_node = self._build(X[right_mask], y[right_mask], active_features, new_protected)

            return RLTNode(
                is_leaf=False, beta=beta, threshold=best_thr,
                left=left_node, right=right_node
            )

    def _make_leaf(self, y):
        if self.task == "regression":
            return RLTNode(is_leaf=True, value=float(np.mean(y)))
        else:
            p1 = float(np.mean(y))
            return RLTNode(is_leaf=True, value=p1)  # proba classe 1

    # --------------------------
    # API fit / predict
    # --------------------------
    def fit(self, X, y):
        X = np.asarray(X); y = np.asarray(y)
        self.n_features_ = X.shape[1]
        active_features = list(range(self.n_features_))
        protected = set()
        self.root = self._build(X, y, active_features, protected)
        return self

    def _predict_one(self, x, node):
        if node.is_leaf:
            return node.value
        if node.beta is not None:
            z = float(np.dot(x, node.beta))
            if z <= node.threshold:
                return self._predict_one(x, node.left)
            else:
                return self._predict_one(x, node.right)
        else:
            if x[node.feature] <= node.threshold:
                return self._predict_one(x, node.left)
            else:
                return self._predict_one(x, node.right)

    def predict(self, X):
        X = np.asarray(X)
        preds = np.array([self._predict_one(x, self.root) for x in X])
        if self.task == "classification":
            return (preds >= 0.5).astype(int)
        return preds

    def predict_proba(self, X):
        # proba classe 1
        X = np.asarray(X)
        p1 = np.array([self._predict_one(x, self.root) for x in X])
        return np.vstack([1-p1, p1]).T


class RLTForest:
    """
    Forêt RLT = bagging de RLTTree (comme Random Forest).
    """
    def __init__(
        self,
        n_estimators=100,
        task="regression",
        **tree_kwargs
    ):
        self.n_estimators = n_estimators
        self.task = task
        self.tree_kwargs = tree_kwargs
        self.trees = []
        self.rng = np.random.default_rng(tree_kwargs.get("random_state", None))

    def fit(self, X, y):
        X = np.asarray(X); y = np.asarray(y)
        n = X.shape[0]
        self.trees = []

        for _ in range(self.n_estimators):
            # bootstrap
            idx = self.rng.integers(0, n, size=n)
            Xb, yb = X[idx], y[idx]

            tree = RLTTree(task=self.task, **self.tree_kwargs)
            tree.fit(Xb, yb)
            self.trees.append(tree)

        return self

    def predict(self, X):
        X = np.asarray(X)
        preds = np.array([t.predict(X) for t in self.trees])
        if self.task == "classification":
            # vote majoritaire
            return (preds.mean(axis=0) >= 0.5).astype(int)
        else:
            return preds.mean(axis=0)

    def predict_proba(self, X):
        X = np.asarray(X)
        p1 = np.array([t.predict_proba(X)[:,1] for t in self.trees]).mean(axis=0)
        return np.vstack([1-p1, p1]).T
```

```python
# ============================================================
# MODELING + EVALUATION — Baselines (sklearn) + RLT (R package)
# Compatible avec tes variables:
# X_train_auto, X_test_auto, y_train_auto, y_test_auto
# X_train_ozone, X_test_ozone, y_train_ozone, y_test_ozone
# ============================================================

import numpy as np
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import roc_auc_score, f1_score, recall_score, precision_score, accuracy_score

# -----------------------------
# Utils d'évaluation
# -----------------------------
def eval_regression(y_true, y_pred, name="model"):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    print(f"[{name}] MSE={mse:.4f} | RMSE={rmse:.4f} | R2={r2:.4f}")
    return {"mse": mse, "rmse": rmse, "r2": r2}

def eval_classification(y_true, y_score, thr=0.5, name="model"):
    # y_score = proba classe 1
    y_pred = (y_score >= thr).astype(int)
    auc = roc_auc_score(y_true, y_score)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    prec = precision_score(y_true, y_pred, zero_division=0)
    acc = accuracy_score(y_true, y_pred)
    print(f"[{name}] AUC={auc:.4f} | F1={f1:.4f} | Recall={rec:.4f} | Precision={prec:.4f} | Acc={acc:.4f}")
    return {"auc": auc, "f1": f1, "recall": rec, "precision": prec, "acc": acc}

# ============================================================
# (A) BASELINES — Auto MPG (régression)
# ============================================================

print("========== BASELINES — AUTO MPG ==========")

rf_reg = RandomForestRegressor(
    n_estimators=500, random_state=42, n_jobs=-1
)
et_reg = ExtraTreesRegressor(
    n_estimators=500, random_state=42, n_jobs=-1
)

rf_reg.fit(X_train_auto, y_train_auto)
et_reg.fit(X_train_auto, y_train_auto)

pred_rf_auto = rf_reg.predict(X_test_auto)
pred_et_auto = et_reg.predict(X_test_auto)

eval_regression(y_test_auto, pred_rf_auto, "RandomForestRegressor")
eval_regression(y_test_auto, pred_et_auto, "ExtraTreesRegressor")

# ============================================================
# (B) BASELINES — Ozone (classification binaire rare)
# ============================================================

print("\n========== BASELINES — OZONE ==========")

rf_clf = RandomForestClassifier(
    n_estimators=500, random_state=42, n_jobs=-1,
    class_weight="balanced"  # utile car classe 1 rare
)
et_clf = ExtraTreesClassifier(
    n_estimators=500, random_state=42, n_jobs=-1,
    class_weight="balanced"
)

rf_clf.fit(X_train_ozone, y_train_ozone)
et_clf.fit(X_train_ozone, y_train_ozone)

proba_rf_ozone = rf_clf.predict_proba(X_test_ozone)[:, 1]
proba_et_ozone = et_clf.predict_proba(X_test_ozone)[:, 1]

eval_classification(y_test_ozone, proba_rf_ozone, name="RandomForestClassifier")
eval_classification(y_test_ozone, proba_et_ozone, name="ExtraTreesClassifier")

# ============================================================
# (C) RLT — via package R officiel (CRAN) avec rpy2
# ============================================================
# Le package RLT implémente exactement RLT (Zhu et al. 2015)
# Usage : RLT(x, y, model="regression"/"classification",
#             reinforcement=TRUE, ntrees=100, combsplit=2/3, combsplit.th=0.25, importance=TRUE)
# Predict : predict(fit, testx)$Prediction
# cf. doc CRAN RLT 3.2.6

!pip -q install rpy2

%load_ext rpy2.ipython

import rpy2.robjects as ro
from rpy2.robjects import numpy2ri
numpy2ri.activate()

# ---- envoyer les matrices python vers R ----
ro.globalenv["trainx_auto"] = X_train_auto
ro.globalenv["testx_auto"]  = X_test_auto
ro.globalenv["trainy_auto"] = y_train_auto
ro.globalenv["testy_auto"]  = y_test_auto

ro.globalenv["trainx_oz"] = X_train_ozone
ro.globalenv["testx_oz"]  = X_test_ozone
ro.globalenv["trainy_oz"] = y_train_ozone
ro.globalenv["testy_oz"]  = y_test_ozone
```
